```python
class MyGradientLinearRegression(MyLinearRegression):
    def __init__(self,**kwargs):
        super().__init__(**kwargs)
        self.w = None
        
    def fit(self, X, y, lr = 0.01, max_iter = 100):
        n, k = X.shape
        
        if self.w is None:
            self.w = np.random.randn(k + 1 if self.bias else k)
        
        X_train = np.hstack((X, np.ones((n,1)))) if self.bias else X
        
        self.losses = []
        
        for iter_n in range(max_iter):
            y_pred = self.predict(X)
            self.losses.append(mean_squared_error(y_pred, y))
            
            grad = self._calc_gradient(X_train, y, y_pred)
            
            assert grad.shape == self.w.shape, f'gradient shape {grad.shape} is not equal weight shape {self.w.shape}'
            self.w -= lr * grad
            
        return self
    
    def _calc_gradient(self, X, y, y_pred):
        grad = 2 * (y_pred - y)[:, np.newaxis] * X
        grad = grad.mean(axis = 0)
        return grad
    
    def get_losses(self):
        return self.losses
    
```

```python
def linear_expression(x):
    return 5 * x + 6
```

```python
objects_num = 50
X = np.linspace(-5, 5, objects_num)
y = linear_expression(X) + np.random.randn(objects_num) * 5

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5)
```

```python
regressor = MyGradientLinearRegression(bias = True)

l = regressor.fit(X_train[:, np.newaxis], y_train, max_iter =100).get_losses()
predictions = regressor.predict(X_test[:, np.newaxis])
w = regressor.get_weights()
```

```python
plt.figure(figsize = (10,10))
plt.plot(X, linear_expression(X), label = 'real', c = 'g')

plt.scatter(X_train, y_train, label = 'train')
plt.scatter(X_test, y_test, label = 'test')

plt.plot(X, regressor.predict(X[:, np.newaxis]), label = 'prediction', c = 'r')

plt.grid(alpha = 0.2)
plt.legend()
plt.show()
```

```python
# plot GD learning

plt.figure(figsize=(10,7))

plt.plot(l)

plt.title('Gradient Descent learning')
plt.xlabel('iterations')
plt.ylabel('loss')
plt.ylim(bottom = 0)
plt.grid(alpha = 0.2)

plt.show()
```

```python
# SGD mini-batch samples

class MySGDLinearRegression(MyGradientLinearRegression):
    def __init__(self, samples = 10, **kwargs):
        super().__init__(**kwargs)
        self.w = None
        self.samples = samples
        
    def _calc_gradient(self, X, y, y_pred):
        i = np.random.choice(np.arange(X.shape[0]), size = self.samples, replace = False)
        
        grad = 2 * (y_pred[i] - y[i])[:, np.newaxis] * X[i]
        grad = grad.mean(axis = 0)
        
        return grad
```

```python
regressor = MySGDLinearRegression(bias = True)

l = regressor.fit(X_train[:, np.newaxis], y_train, max_iter=100).get_losses()

predictions = regressor.predict(X_test[:, np.newaxis])
w = regressor.get_weights()
```

```python
plt.figure(figsize=(10,10))

plt.plot(X, linear_expression(X), label = 'real', c = 'g')

plt.scatter(X_train, y_train, label = 'train')
plt.scatter(X_test, y_test, label = 'test')

plt.plot(X, regressor.predict(X[:, np.newaxis]), label = 'predicted', c = 'b')


plt.grid(alpha = 0.2)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))

plt.plot(l)
plt.xlabel('iterations')
plt.ylabel('loss')
plt.ylim(bottom = 0)
plt.grid(alpha = 0.2)

plt.title('GD loss over iterations')
plt.show()
```
