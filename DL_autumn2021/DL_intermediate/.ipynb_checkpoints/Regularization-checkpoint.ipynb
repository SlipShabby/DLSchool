```python
import numpy as np
import pandas as pd
import scipy.linalg as sla
import matplotlib.pyplot as plt
import seaborn as sns

import import_ipynb
from LinearRegression import MyLinearRegression
from GradientDescent import MySGDLinearRegression
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.datasets import make_blobs

from sklearn.preprocessing import StandardScaler
```

```python
def linear_expression(x):
    return 5 * x + 6

objects_num = 50
X = np.linspace(-5, 5, objects_num)
y = linear_expression(X) + np.random.randn(objects_num) * 5

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5)
```

```python
# Ridge (L2-regularization) regression
```

```python
reg = Ridge(alpha=1).fit(np.hstack((X,X,X))[:, np.newaxis], np.hstack((y,y,y)))
np.append(reg.coef_, reg.intercept_)
```

```python
class MyRidgeRegression(MyLinearRegression):
    def __init__(self, alpha=0.1, **kw):
        super().__init__(**kw)
        self.alpha = alpha
        
    def fit(self, X, y):
        n, m = X.shape
        X_train = X
        
        if self.bias:
            X_train = np.hstack((X, np.ones((n,1))))
            
        lambdaI = self.alpha * np.eye(X_train.shape[1])
        if self.bias:
            lambdaI[-1,-1] = 0
            
        self.w = np.linalg.inv(X_train.T @X_train + lambdaI) @ X_train.T @y
        
        return self
    
    def get_weight(self):
        return self.w
    
    
```

```python
alpha = 1.0
regressor = MyRidgeRegression(alpha = alpha).fit(X_train[:,np.newaxis], y_train)
```

```python
sklearn_reg = Ridge(alpha).fit(X_train[:, np.newaxis], y_train)
assert np.allclose(regressor.get_weight(), np.append(sklearn_reg.coef_, sklearn_reg.intercept_))
regressor.get_weight()
```

```python
# SGD 

class MySGDRidgeRegression(MySGDLinearRegression):
    def __init__(self, alpha = 1.0, **kw):
        super().__init__(**kw)
        self.w = None
        self.alpha = alpha
        
    def _calc_gradient(self, X, y, y_pred):
        i = np.random.choice(np.arange(X.shape[0]), size = self.samples, replace = False)
        
        lambdaI = self.alpha * np.eye(self.w.shape[0])
        if self.bias:
            lambdaI[-1, -1] = 0
            
        grad = 2 * (X[i].T @ X[i] / self.samples + lambdaI)@self.w
        grad -= 2 * X[i].T @ y[i] / self.samples
        
        return grad
```

```python
plt.figure(figsize=(10,7))
plt.plot(X, linear_expression(X), label = 'real', c = 'g')

plt.scatter(X_train, y_train, label = 'train')
plt.scatter(X_test, y_test, label = 'test')

plt.plot(X, regressor.predict(X[:, np.newaxis]), label = 'predicted', c = 'b')

plt.grid(alpha = 0.2)
plt.legend()
plt.show()
```

```python
regressor = MySGDRidgeRegression(alpha = 1, samples = 20).fit(X[:,np.newaxis], y, max_iter=1000, lr = 0.01)
l = regressor.get_losses()
regressor.get_weights()
```

```python
def soft_sign(x, eps=1e-7):
    if abs(x) > eps:
        return np.sign(x)
    return x / eps

np_soft_sign = np.vectorize(soft_sign)
```

```python
class MySGDLasso(MySGDLinearRegression):
    def __init__(self, alpha=1.0, **kw):
        super().__init__(**kw) 
        self.alpha = alpha

    def _calc_gradient(self, X, y, y_pred):

        inds = np.random.choice(np.arange(X.shape[0]), size=self.samples, replace=False)

        signw = np_soft_sign(self.w)
        if self.bias:
            signw[-1] = 0

        grad = X[inds].T @ (y_pred[inds] - y[inds])[:, np.newaxis] / self.samples
        grad += self.alpha * signw[:, np.newaxis]

        return grad.flatten()
```

```python
regressor = MySGDLasso(alpha=1, samples=4).fit(X[:, np.newaxis], y, max_iter=1000, lr=0.01)
l = regressor.get_losses()
regressor.get_weights()
```

```python
sklearn_reg = Lasso().fit(X[:, np.newaxis], y)
np.append(sklearn_reg.coef_, sklearn_reg.intercept_)
```

```python
names = data.columns
data.head()
```

```python
sns.pairplot(data, hue = 'target')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[names[:-1]], data[names[-1]])
```

```python
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, f1_score
```

```python
cls = make_pipeline(StandardScaler(), LogisticRegression(C=2))

cls = cls.fit(X_train.to_numpy(), y_train)
preds_train = cls.predict(X_train)
```

```python
accuracy_score(preds_train, y_train), f1_score(preds_train,y_train, average='macro')
```

```python
preds_test = cls.predict(X_test)

accuracy_score(preds_test, y_test), f1_score(preds_test, y_test, average='macro')
```
